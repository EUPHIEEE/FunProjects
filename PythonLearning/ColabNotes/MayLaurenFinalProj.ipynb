I've modified this, but the code uses [Networkx research ](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.centrality.eigenvector_centrality.html#networkx.algorithms.centrality.eigenvector_centrality ) to find which actors in the network are the most important with a given phrase...

```python

nest_asyncio.apply()

c = twint.Config()

phrase = "#MAMAVote"
c.Search = phrase
c.Limit = 2000
c.Store_csv = True
c.Output = "tweetx.csv"

twint.run.Search(c)

tweets=pd.read_csv("tweetx.csv")
tweets

tweets=tweets[tweets['reply_to'] !='[]']
tweets['reply_to']=tweets['reply_to'].str[1:]
a=(tweets.assign(list=tweets["reply_to"].str.split(", ")).explode("list"))

c=a[a.list.str.contains("'screen")]

c['list']=c['list'].str[17:]
c['list']=c['list'].str[:-1]

c.head()
c.to_csv(r'artru.csv')

G=nx.from_pandas_edgelist(c,source='username', target='list', create_using=nx.DiGraph())
ec = nx.eigenvector_centrality(G, tol=1e-03)
cc = nx.closeness_centrality(G)
sc1 = sorted(ec.items(), key=lambda x: x[1], reverse=True)
sc2 = sorted(cc.items(), key=lambda x: x[1], reverse=True)

nx.draw_networkx(G, with_labels=False, node_size=80, arrowsize=5, alpha=0.3, width=0.3)

print("10 most influential users in" + phrase +" are:\n\nEigenvector Centrality:\n")
sc1[:10]

print("Closeness Centrality:\n")
sc2[:10]
```
